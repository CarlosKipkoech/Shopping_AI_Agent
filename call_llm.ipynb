```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
```

```python
llm = ChatGoogleGenerativeAI(model="gemma-4-31b-it")

response = llm.invoke([
    ["system","Respond in just one line"],
    ["human","How many moons does jupiter have?"]])

print(response.text)
```

```python
prompt = "Write a nice and creative tagline for my Applied AI  finance software company. respond in one line"

llm = ChatGoogleGenerativeAI(model="gemma-4-31b-it", temperature=0)
response=llm.invoke(prompt)
print(response.text)
```
