```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
 
```

```python
extraction_prompt = f"""You are a medical report extraction assistant. 

Your task is to carefully read the medical report and extract the important clinical information in a clean, structured format. 
Instructions:
- Extract patient details.
- Extract report date.
- Group results under their respective sections.
- For each test, extract:
  - test name
  - measured value
  - unit
  - reference/normal range
  - interpretation (Normal, High, or Low)
  Blood report: {blood_report}
"""
extraction_response = llm.invoke(extraction_prompt)
extracted_values = extraction_response.text
print("=== STAGE 1: EXTRACTED VALUES ===")
print(extracted_values)
```

```python
diet_prompt= f""" You are a nutritionist.

Using the medical report:
- Write a short health summary
- Explain the condition in simple language
- Give practical diet advice
- Suggest foods to avoid and foods to eat more

Keep it short of upto just 100 words, simple, and practical.
Avoid medical jargon.

Report:
{extracted_values}
"""
diet_response = llm.invoke(diet_prompt)
report = diet_response.text

print("========DIETARY REPORT========")
print(report)
```
