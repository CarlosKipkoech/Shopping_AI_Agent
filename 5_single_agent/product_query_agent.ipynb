```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain.tools import tool
from langchain.agents import create_agent

load_dotenv()

```

```python
PRODUCTS = {
    "wireless earbuds": {
        "price": 89.99,
        "description": "Bluetooth 5.3, noise cancellation, 24-hour battery life"
    },

    "gaming mouse": {
        "price": 59.99,
        "description": "RGB lighting, adjustable DPI, ergonomic lightweight design"
    },

    "smart speaker": {
        "price": 129.99,
        "description": "Voice assistant support, immersive sound, WiFi connectivity"
    },

    "portable ssd": {
        "price": 149.99,
        "description": "1TB storage, ultra-fast transfer speeds, compact aluminum body"
    },

    "4K monitor": {
        "price": 329.99,
        "description": "27-inch UHD display, HDR support, slim bezel design"
    },

    "fitness tracker": {
        "price": 79.99,
        "description": "Heart-rate monitoring, sleep tracking, waterproof build"
    },

    "mechanical keyboard": {
        "price": 139.99,
        "description": "Hot-swappable switches, RGB backlight, tactile typing experience"
    },

    "laptop cooling pad": {
        "price": 39.99,
        "description": "Dual cooling fans, adjustable height, silent airflow system"
    }
}
```

```python
# Define tool 
@tool
def get_product(name: str) -> str:
    """ Look up a product by name and return its price, rating, stock, and description """

    p = PRODUCTS.get(name.lower())
    if not p: 
        return f" Product not found. Available products are: {','.join(PRODUCTS)}"
    return str(p)

#creating an Agent.

llm = ChatGroq(model="llama-3.3-70b-versatile",temperature=0)

agent = create_agent(
    llm,
    tools = [get_product],
    system_prompt = "You are a helpful product assistant for an online tech store"
)

```

```python
def ask (question: str):
    result = agent.invoke({"messages": [{"role": "user", "content":question}]})
    print(result["messages"][-1].content)
```

```python
ask("what is the  total price of any keyboard and a storage device")
```
