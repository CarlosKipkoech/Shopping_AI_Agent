```python
import chromadb

client = chromadb.Client()

collection = client.create_collection("news_article")
```

```python
collection.add(

    ids= ["id1","id2","id3","id4"],
    documents = [
    "Global markets surged today after technology companies reported stronger quarterly earnings.",
    "Heavy rainfall caused flooding overnight, forcing hundreds of residents evacuations yesterday.",
    "Scientists discovered promising renewable energy materials improving battery efficiency and sustainability.",
    "Local football team celebrated championship victory after dramatic penalty shootout final."
]
)
```

```python
results = collection.query(

    query_texts = ["This is a query related to Elon musk"],
    n_results = 2
    
)
print(results)
```
